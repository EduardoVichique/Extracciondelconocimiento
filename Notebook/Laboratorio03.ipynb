```python
import pandas as pd
import numpy as np
```

```python
import random
```

```python
from datetime import datetime, timedelta
```

```python
# =========================================
# 2. CONFIGURACIÓN
# =========================================

random.seed(42)
np.random.seed(42)

n_publicaciones = 1000000
```

```python
# 3. DATOS BASE
influencers = [
    "fit_carlos",
    "gym_diaries",
    "runner_mx",
    "yoga_with_sofia",
    "crossfit_pro",
    "musculo_total",
    "vida_fitness",
    "keto_mexico"
]

tipos_post = [
    "foto",
    "reel",
    "historia",
    "video"
]
```

```python
# 4. GENERAR EL DATASET
dataset = pd.DataFrame({
    
    "influencer": [
        random.choice(influencers)
        for _ in range(n_publicaciones)
    ],
    
    "likes": [
        random.randint(100, 50000)
        for _ in range(n_publicaciones)
    ],
    
    "comentarios": [
        random.randint(0, 5000)
        for _ in range(n_publicaciones)
    ],
    
    "fecha": [
        datetime(2025,1,1) + timedelta(days=random.randint(0,365))
        for _ in range(n_publicaciones)
    ],
    
    "tipo_post": [
        random.choice(tipos_post)
        for _ in range(n_publicaciones)
    ]
})
```
