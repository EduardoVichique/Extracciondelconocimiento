```python
#Importammos la libreria Panda
import pandas as pd
```

```python
#Conjunto de datos base para la generacion de BIGDATA
datos = {
    "usuario": ["ana","pepe","luis", "rosita","carolina","paco",],
    "edad":[14,22,35,40,62,70],
    "ciudad" : ["cuitlahuac","yanga", "atoyac", "paraje", "fortin", "cuichapa"],
    "compras":[5,10,1,2,15,3],
    "activo":[True,True,True,False,True,False],
    "fecha" : ["2026-01-10","2026-05-21", "2026-02-10", "2026-03-11", "2026-04-15", "2026-07-25"]
}
```

```python
#Creamos el dataframe
dfDatos = pd.DataFrame(datos)
```

```python
dfDatos
```

```python
#Cantidad de registros a generar aleatoriamente
cantidad = 1000000
```

```python
import random
```

```python
#Crear el dataFrame de 1000000 de registros
dfMillon = pd.DataFrame(
    {
        "nombre": [random.choice(dfDatos["usuario"]) for _ in range (cantidad)],
        "edad": [random.randint(18,80)for _ in range (cantidad)],
        "ciudad": [random.choice(dfDatos["ciudad"])for _ in range(cantidad)],
        "compras": [random.randint(1,10)for _ in range (cantidad)],
        "activo": [random.choice(dfDatos["activo"]) for _ in range (cantidad)],
        "fecha": [random.choice(dfDatos["fecha"]) for _ in range (cantidad)],


    }
)
```
